```python
import matplotlib.pyplot as plt

numeric_columns = ['avg_rating', 'num_of_reviews', 'rating']

for col in numeric_columns:
    plt.figure(figsize=(8, 4))
    plt.hist(df_fede[col], bins=20, color='skyblue')
    plt.title(f'Histograma de {col}')
    plt.xlabel(col)
    plt.ylabel('Frecuencia')
    plt.show()
```

```python
plt.scatter(df_fede['avg_rating'], df_fede['num_of_reviews'])
plt.xlabel('Calificación Promedio')
plt.ylabel('Número de Reseñas')
plt.title('Gráfico de Dispersión entre Calificación Promedio y Número de Reseñas')
plt.show()
```

```python
plt.scatter(df_fede['num_of_reviews'], df_fede['rating'])
plt.xlabel('Número de Reseñas')
plt.ylabel('Calificación Individual')
plt.title('Gráfico de Dispersión entre Número de Reseñas y Calificación Individual')
plt.show()
```

```python
correlation_matrix = df_fede[numeric_columns].corr()
```

```python
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Matriz de Correlación')
plt.show()
```

```python
import pandas as pd

# Ejemplo de DataFrame con una columna 'category'
data = {'category': ["['Korean restaurant']", "['Italian restaurant']"]}
df = pd.DataFrame(data)

# Utiliza str.extract() con una expresión regular para extraer "restaurant"
df['restaurant_type'] = df['category'].str.extract(r"'(.*?)'")

# Mostrar el DataFrame resultante
print(df)
```

```python
import pandas as pd

# Ejemplo de DataFrame con una columna 'time'
data = {'time': ['2021-01-06 05:12:07.056000000', '2021-01-07 08:30:15.123000000']}
df = pd.DataFrame(data)

# Convierte la columna 'time' al formato datetime
df['time'] = pd.to_datetime(df['time'])

# Extrae la fecha y crea una nueva columna 'date'
df['date'] = df['time'].dt.date

# Muestra el DataFrame resultante
print(df)
```

```python
# Convierte la columna 'time' al formato datetime
df_fede['time'] = pd.to_datetime(df_fede['time'])
```

```python
# Extrae la fecha y crea una nueva columna 'date'
df_fede['date'] = df_fede['time'].dt.date
```

```python
# Obtén la cantidad total de filas en el DataFrame original
total_filas = len(df_fede)

# Calcula el punto medio para dividir el DataFrame en dos mitades
mitad = total_filas // 2

# Divide el DataFrame en dos DataFrames nuevos
df_fede_primera_mitad = df_fede.iloc[:mitad]  # Primera mitad de las filas
df_fede_segunda_mitad = df_fede.iloc[mitad:]  # Segunda mitad de las filas
```

```python
# Supongamos que tienes el DataFrame df_fede_primera_mitad y deseas guardarlo como un archivo CSV
df_fede_primera_mitad.to_csv('gmaps_primera_mitad.csv', index=False)
```
